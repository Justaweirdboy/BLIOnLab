```kotlin


// 1. CSV-k
val rf  = DataFrame.readCSV("C:\\Dev\\smile\\predictions_RandomForest.csv")
val gb  = DataFrame.readCSV("C:\\Dev\\smile\\predictions_GradientBoosting.csv")
val rid = DataFrame.readCSV("C:\\Dev\\smile\\predictions_Ridge.csv")

// 2. átnevezés + join "row" szerint
val merged = rf.rename("predicted" to "rf")
    .innerJoin(                       // <-- nincs 'on ='
        gb.rename("predicted" to "gb").remove("actual"),
        "row"
    )
    .innerJoin(
        rid.rename("predicted" to "ridge").remove("actual"),
        "row"
    )
    .sortBy("row")

plot(merged.take(100)) {
    line { x("row"); y("actual"); color = Color.RED;    }
    line { x("row"); y("rf");     color = Color.BLUE;  }
    line { x("row"); y("gb");     color = Color.GREEN;  }
    line { x("row"); y("ridge");  color = Color.ORANGE; }
    layout {
        title = "Actual vs. Predicted (first 100 rows)"
        xAxisLabel = "row"
        yAxisLabel = "cloudTransparency"
        size = 9000 to 5000
    }
}
```
